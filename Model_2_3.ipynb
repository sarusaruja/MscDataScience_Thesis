```python
causal_graph = """
digraph {
    risk_rate;
    target_default;
    last_amount_borrowed;
    credit_limit;
    income;
    n_bankruptcies;
    n_defaulted_loans;
    income -> target_default;
    U[label="Unobserved Confounders"];
    n_defaulted_loans -> target_default;
    credit_limit -> target_default;
    U -> risk_rate;
    U -> target_default;
    U -> n_defaulted_loans;
    n_bankruptcies -> risk_rate;

    risk_rate -> target_default;
    last_amount_borrowed -> risk_rate;
}

"""
```

```python
from dowhy import CausalModel
```

```python
credit_data_causal= credit_data[['risk_rate', 'target_default', 'last_amount_borrowed', 'credit_limit', 'income', 'n_bankruptcies', 'n_defaulted_loans']
].copy()
```

```python
credit_data_causal
```

```python
model= CausalModel(
        data = credit_data_causal,
        graph=causal_graph.replace("\n", " "),
        treatment='risk_rate',
        outcome='target_default')
model.view_model()
```

```python
# from IPython.display import Image, display
# display(Image(filename="causal_model.png"))
```

```python
#Identify the causal effect
estimands = model.identify_effect()
print(estimands)
```

#Model 2: Causal Features + XGBoost

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns

causal_data = credit_data_causal
label_encoder = LabelEncoder()
causal_data['target_default'] = label_encoder.fit_transform(causal_data['target_default'])

train, test = train_test_split(causal_data, test_size=0.2, random_state=200)

# Feature Selection
features = ['risk_rate','last_amount_borrowed', 'credit_limit', 'income', 'n_bankruptcies', 'n_defaulted_loans']

xgb_model = xgb.XGBClassifier()
xgb_model.fit(train[features], train['target_default'])

# Prediction on test data
y_pred = xgb_model.predict(test[features])

accuracy = accuracy_score(test['target_default'], y_pred) * 100
print("Total Accuracy:", accuracy, "%")

feature_importance = xgb_model.feature_importances_

# Plotting feature importance
plt.figure(figsize=(10, 6))
plt.barh(features, feature_importance)
plt.xlabel('Feature Importance')
plt.title('Feature Importance Plot')
plt.show()
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns

causal_data = credit_data_causal
label_encoder = LabelEncoder()
causal_data['target_default'] = label_encoder.fit_transform(causal_data['target_default'])

# Train-test split
X = causal_data.drop('target_default', axis=1)
y = causal_data['target_default']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=200, stratify=y)

# Feature Selection
features = ['risk_rate','last_amount_borrowed', 'credit_limit', 'income', 'n_bankruptcies', 'n_defaulted_loans']

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'n_estimators': [100, 200, 300],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'reg_alpha': [0, 0.1, 0.5, 1.0],
    'reg_lambda': [0, 0.1, 0.5, 1.0],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 0.1, 0.5, 1.0]
}

xgb_model = xgb.XGBClassifier()

# Hyperparameter tuning
random_search = RandomizedSearchCV(estimator=xgb_model, param_distributions=param_grid,
                                   n_iter=10, scoring='accuracy', cv=5, verbose=1, random_state=42)

random_search.fit(X_train[features], y_train)

# Get best estimator
best_xgb_model = random_search.best_estimator_

# Predict on test data
y_pred = best_xgb_model.predict(X_test[features])

accuracy = accuracy_score(y_test, y_pred) * 100
print("Total Accuracy:", accuracy, "%")

print("Best Parameters:", random_search.best_params_)
```
